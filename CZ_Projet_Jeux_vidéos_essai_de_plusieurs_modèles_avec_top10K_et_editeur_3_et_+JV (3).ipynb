**Modelisationdu dataset : "dataset_jeux_par_item V2" top 10K et avec les éditeurs de 3 ou plus jeux vidéos. **

```python
#Librairies de base 
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#lecture du dataset 
df=pd.read_csv("dataset_jeux_par_item V2.csv")
df.head(10)
```

```python
#colonne année mise en format datetime et remplacement des valeurs manquantes par l'année 1900 
#suffisamment ancienne pour ne pas affecter les résultats d'analyse
from datetime import datetime as dt 
df['annee']=pd.to_datetime(df['annee'], errors="coerce").fillna(pd.Timestamp("1900"))
df['annee'] = df['annee'].dt.year
df["annee"].info()
```

```python
#prise du top 10K 
df_top10K=df.head(10000)
```

```python
#selection des éditeurs ayant 3 ou plus de jeux vidéos 
counts = df_top10K['editeur'].value_counts()
selected_publishers = counts[counts >= 3].index
df_top10K= df_top10K[df_top10K['editeur'].isin(selected_publishers)]
```

```python
#info du dataset obtenu
df_top10K.info()
```

```python
#split de la colonne igdb_age_rating pour ne garder que l'ESRB 
df_top10K['ESRB'] = df_top10K['igdb_age_rating_class'].str.split(',').str[0].apply(lambda x: x.strip() if str(x).startswith('ESRB') else pd.NaT)
```

```python
#verif des colonnes 
df_top10K.columns
```

```python
#suppression de colonnes pour ne garder qu'un dataset de base (Genre, nom_plateforme, nom_fabricant, editeur, developpeur, année, ttl_sales, dev_country et ESRB)
colonnes_a_supprimer=['metacritic_summary','metacritic_user_score', 'metacritic_score',
       'jvc_summmary','igdb_id', 'igdb_category','igdb_franchise', 'igdb_platforms', 'igdb_genres', 'igdb_game_modes', 'igdb_summary', 'igdb_themes','igdb_url', 'igdb_age_content_descriptions', 'igdb_age_rating_category',
       'igdb_age_rating_class', 'igdb_developer_name',
       'igdb_developer_country', 'igdb_publisher_name',
       'igdb_publisher_country', 'igdb_websites_category', 'igdb_websites_url',
       'twitch_id', 'jvc_url','jeu','jvc_user_score', 'jvc_url', 'mixed_user_score','igdb_follows', 'igdb_hypes',
       'igdb_franchise','igdb_rating','igdb_total_rating','twitch_streams_count']
df_top10K = df_top10K.drop(columns=colonnes_a_supprimer)
```

```python
df_top10K.info()
```

```python
# encoder les variables catégorielles
cat_vars = ["Genre", "nom_plateforme", "nom_fabricant", "editeur", "developpeur", "dev_country", "ESRB"]

df_top10K_encoded = pd.get_dummies(df_top10K, columns=cat_vars)
```

```python
#verif valeurs manquantes
df_top10K_encoded.isna().sum().sum()
```

```python
#préparation pour séparation des données 
df_top10K_encoded=df_top10K_encoded.drop(['ttl_sales'], axis=1)
y=df_top10K.ttl_sales
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(df_top10K_encoded,y, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test
```

```python
#entrainement  du modèle de régression par vecteurs de support (SVR - Support Vector Regression) 
#et optimisation des hyperparamètres du modèle à l'aide de la validation croisée en grille (GridSearchCV)
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error


# parametres GridSearchCV
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'degree': [2, 3, 4]
}

# SVR
svr = SVR()

#  GridSearchCV
grid_search = GridSearchCV(svr, param_grid=param_grid, cv=5, n_jobs=-1)

# Fit GridSearchCV sur les données d'entrainement
grid_search.fit(X_train, y_train)

# Prediction
y_pred = grid_search.best_estimator_.predict(X_test)

# Evaluation 
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE of SVR model: {:.2f}".format(rmse))
```

```python
#l'erreur quadratique moyenne entre les valeurs prédites et les vraies valeurs est d'environ 2.96 Millions (échelle de la variable cible : ttls_sales)
```

```python
#entrainement d'un modèle de réseau de neurones artificiels (ou "deep learning") pour effectuer une tâche de régression.
import tensorflow as tf
from tensorflow import keras

#model tensorflow
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1)
])

# Compilation
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Fit
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Prediction
y_pred = model.predict(X_test)

# Evaluation
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE of neural network model: {:.2f}".format(rmse))
```

```python
#entrainement  d'un modèle de régression de Boosting d'arbres de décision.
import xgboost as xgb

#objet DMatrix pour les données d'entraînement et de test
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

#paramètres du modèle
params = {'max_depth': 3, 'eta': 0.1, 'objective': 'reg:squarederror'}

# Entraîner le modèle
model = xgb.train(params, dtrain, num_boost_round=100)

#prédictions sur les données de test
y_pred = model.predict(dtest)

# Évaluer le modèle en utilisant la RMSE
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE:", rmse)
```

```python
#entrainement par un modèle de regression qui utilise un arbre de décision 
from sklearn.tree import DecisionTreeRegressor
modelregressor=DecisionTreeRegressor()
MR=modelregressor.fit(X_train, y_train)
print(MR.score(X_train,y_train))
print(MR.score(X_test, y_test))
```

```python
#Le score correspond au coeff de détermination (R2) :mesurant la proportion de variance totale des données de la target (ttl_sales)
#Ici Le modèle explique 83% de la variance des données d'entrainement et 33 % de la variance des données test
```

```python
#entrainement du modèle qui utilise la régression Ridge 
#(technique de régularisation qui ajoute une pénalité L2 à la fonction de coût pour limiter la magnitude des coefficients de régression 
#et éviter le surajustement du modèle)
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error

# modèle de régression ridge (hyperparamètre alpha qui contrôle l'intensité de la pénalité L2)
ridge = Ridge(alpha=1.0)

# entraînemeht du  modèle 
ridge.fit(X_train, y_train)

# évaluation perf 
y_pred = ridge.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)
```

```python
#les prédictions en moyenne ont une erreur absolue de 0.9576 unité de la variable cible ttls_sales par rapport aux valeurs réelles des données de test.
```
