```python
#librairie
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset_jeux_par_item V2.csv")
df.head(10)
```

```python
df.isna().sum()
```

```python
#colonne annÃ©e mise en format datetime et remplacement des valeurs manquantes par l'annÃ©e 1900 suffisamment ancienne pour ne pas affecter les rÃ©sultats d'analyse
from datetime import datetime as dt 
df['annee']=pd.to_datetime(df['annee'], errors="coerce").fillna(pd.Timestamp("1900"))
df['annee'].info()
```

```python
df.annee.isna().sum()
```

```python
#variable cible : 
y=df.ttl_sales
#variables explicatives choisies : 
X = df[['editeur', 'nom_plateforme']]
```

```python
#encodage des variables explicatives 
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first', sparse=False)
```

```python
# SÃ©lectionner les colonnes catÃ©gorielles
cat_cols = ['editeur', 'nom_plateforme']

# Encoder les colonnes catÃ©gorielles avec la mÃ©thode "get_dummies"
encoded_data = pd.get_dummies(X, columns=cat_cols)

# Afficher le dataset encodÃ©
print(encoded_data.head())
```

```python
#sÃ©paration du jeu de donnÃ©es : 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(encoded_data,y, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test
```

```python
#Entrainement pa le modÃ¨le Decision Tree Regressor 
from sklearn.tree import DecisionTreeRegressor
modelregressor=DecisionTreeRegressor()
MR=modelregressor.fit(X_train, y_train)
print(MR.score(X_train,y_train))
print(MR.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# PrÃ©dictions sur les donnÃ©es de test
y_pred = MR.predict(X_test)

```

```python
# Erreur quadratique moyenne
mse = mean_squared_error(y_test, y_pred)

# Erreur absolue moyenne
mae = mean_absolute_error(y_test, y_pred)

# Coefficient de dÃ©termination
r2 = r2_score(y_test, y_pred)

# Affichage des scores
print('MSE : {:.2f}'.format(mse))
print('MAE : {:.2f}'.format(mae))
print('RÂ² : {:.2f}'.format(r2))
```
